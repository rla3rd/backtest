```python
import data
import Queue
import pandas as pd
import numpy as np

events = Queue.Queue()
data = data.HistoricDBDataHandler(events, ['AAPL'])
for i in range(0,3650):
    data.update_bars()
    
aapl = pd.DataFrame(data.get_latest_bars('AAPL', N=3650), columns = data.bar_columns)
```

```python
aapl['closepctchg'] = aapl['adjclose'].pct_change(periods=1) * 100
aapl['logreturncalc'] = np.log(aapl['adjclose'].pct_change(periods=1)+1)
aapl['ma200'] = pd.rolling_mean(aapl['adjclose'], 200)
aapl['lowpctchg'] = aapl['low'].pct_change(periods=1)
aapl['highpctchg'] = aapl['high'].pct_change(periods=1)
aapl['lowpctsign'] = np.sign(aapl['lowpctchg'])
aapl['highpctsign'] = np.sign(aapl['highpctchg'])
aapl['closepctsign'] = np.sign(aapl['closepctchg'])
aapl['sumlowsign'] = pd.rolling_sum(aapl['lowpctsign'], 3)
aapl['sumhighsign'] = pd.rolling_sum(aapl['highpctsign'], 3)
aapl['sumclosesign'] = pd.rolling_sum(aapl['closepctsign'], 3)
```
